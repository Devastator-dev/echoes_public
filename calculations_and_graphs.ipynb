Calculation of havoc rover without equipped echoes with emerald of genesis r1

```python
import echoes_main as ech
import additional_functions as addfun
import damage_functions as dmgfun


import numpy as np
import matplotlib
import matplotlib.pyplot as plt


```

```python
average_damages_through_days,percentages_of_max,list_of_days=addfun.simulate_rover_echo_farming(55,6,True,True)
```

```python

Rover_Havoc=ech.Character(ech.set_sse,10825,413,1259,125,5,150,ech.weapon_emerald_of_genesis,'Havoc Rover',ech.main_ha_dmg,0,12,12)
Rover_Havoc._apply_weapon_conditional_bonuses()
damage=[]

Rover_Havoc._total_attack+=(94/100)*Rover_Havoc._base_atk
Rover_Havoc._total_cd+=149
Rover_Havoc._total_cr+=52.5
Rover_Havoc._total_attack+=700
Rover_Havoc._ult_dmg+=58
Rover_Havoc._total_elem_dmg=112

Rover_Havoc.print_character_stats()

for i in range(1000):
    damage.append(dmgfun.calculate_hrover_combo_damage(Rover_Havoc))

print(damage)

fig,ax=plt.subplots()
ax.hist(damage)
ax.set(title='Number of occurances of specific damage with max stats, 1000 tries',xlabel='damage',ylabel='number of occurances')
plt.savefig('Output/Graphs/max damage.png',format='png')

fig,ax=plt.subplots()
ax.plot(np.cumsum(list_of_days),average_damages_through_days,'.')
titstr='Damage during '+str(np.sum(list_of_days))+' days with max theoretical damage and its average for reference, hrover'
ax.set(title=titstr,xlabel='day',ylabel='damage')
ax.plot(10,max(damage),'ro')
ax.plot(10,np.mean(damage),'mo')
plt.savefig('Output/Graphs/'+titstr+'.png')
```
